```python
import pandas as pd
```

#### Data Cleaning 

```python
# import data from webscraping 
df = pd.read_csv('cat_food.csv')
df
```

```python
# start data cleaning
df.columns = ['name','url','Quality_Ingredients','Potential_Allergens','Questionable_Ingrediens','Numeric_Ingredient_name','Numeric_Ingredient','Ingredient_details','pros_cons']
df['Numeric_Ingredient'] = df['Numeric_Ingredient'].str.rstrip('%').apply(lambda x: float(x)/100)
```

```python
index_set = ['name','url','Quality_Ingredients','Questionable_Ingrediens','Ingredient_details','pros_cons','Potential_Allergens']
```

```python
df_pivot = df.pivot(index=index_set,columns='Numeric_Ingredient_name',values='Numeric_Ingredient')
df_pivot = df_pivot.reset_index().rename_axis(None, axis='columns')
new_columns = [i.replace(':','') for i in df_pivot.columns.to_list()]
df_pivot.columns = new_columns
df_pivot['pic'] = df_pivot['name'].apply(lambda x: x+'.jpg')
df_pivot
```

```python
df_nutrients = pd.read_excel('nutrient.xlsx')
df_nutrients
```
